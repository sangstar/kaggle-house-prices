```python
import numpy as np
import json
```

```python
with open('files/to_numerical.txt','r') as f:
    lines = f.readlines()
```

```python
 lines = [line.rstrip() for line in lines]
```

```python
lines
```

```python
to_numerical = ['GarageFinish',
 'Functional',
 'GarageCond',
 'HeatingQC',
 'BsmtCond',
 'PoolQC',
 'KitchenQual',
 'Utilities',
 'ExterCond',
 'BsmtExposure',
 'ExterQual',
 'GarageQual',
 'Fence',
 'LotShape',
 'FireplaceQu',
 'LandSlope',
 'BsmtFinType1',
 'BsmtFinType2',
 'BsmtQual']
```

```python
lines
```

```python
i = 0
ordered_categories = []
while i < len(lines):
    print(i, len(lines))
    line = lines[i]
    split_for_column_name = line.split(":")
    print(split_for_column_name)
    if any([x in to_numerical for x in split_for_column_name]):
        to_add = []
        location_of_match = np.where([x in to_numerical for x in split_for_column_name])[0][0]
        col = split_for_column_name[location_of_match]
        #print(col)
        j = i
        while j+1 != len(lines) and not any([x in to_numerical for x in lines[j+1].split(":")]):
            categorical = lines[j].split("\t")
            to_add.append(categorical[0].replace(" ",''))
            j += 1
        i = j
    to_add = [x for x in to_add if x]
    ordered_categories.append((col, to_add[1:]))
    i += 1
```

```python
ordered_categories = dict(ordered_categories)
```

```python
ordered_categories
```

```python
with open('files/to_numerical_json.json', 'w') as f:
    json.dump(ordered_categories, f)
```
